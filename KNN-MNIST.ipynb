```python
import tensorflow as tf
import datetime
from sklearn import neighbors
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python

trainImage,trainLabel = mnist.train.next_batch(50000)
print(trainImage)
print(trainLabel)
testImage,testLabel = mnist.test.next_batch(10000)
knn = neighbors.KNeighborsClassifier(algorithm = 'auto',leaf_size = 30,n_neighbors=5,weights = 'uniform')  
knn.fit(trainImage,trainLabel)  
match = 0;  
for i in xrange(len(testLabel)):  
    predictLabel = np.argmax(knn.predict(testImage[i])[0])
    print i,' ',  
    print predictLabel,' ',np.argmax(testLabel[i] ) 
    if(predictLabel==np.argmax(testLabel[i])):  
        match += 1  
print 'error rate: '+ str(1-(match*1.0/len(testLabel))) 
```
